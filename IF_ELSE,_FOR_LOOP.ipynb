```python

num = int(input("Enter a number: "))
temp = num
digits = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** digits
    temp = temp // 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python

for num in range(1, 1001):
    digits = len(str(num))
    sum = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        sum = sum + (digit ** digits)
        temp = temp // 10

    if sum == num:
        print(num)
```

```python

n, m = map(int, input("Enter n and m: ").split())


lst = list(range(1, n+1))

result = []

for i in range(0, n, m):
    if i + m <= n:
        group = tuple(lst[i:i+m])
        result.append(group)

print(result)
```
